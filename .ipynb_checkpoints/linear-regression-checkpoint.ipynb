```python
# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
from datetime import timedelta
import matplotlib.dates as mdates
```

```python
df = pd.read_csv(r'C:\Users\M\Desktop\ironhacker\final-project\final1.csv')
df
```

```python
# Prepare the Data for Training
# Use close prices and any additional features as needed (df['date'].values as numeric time)
X = np.array(range(len(df))).reshape(-1, 1)  # Using index as feature; better results with time features
y = df['close'].values  # Target variable

# Split data into train and test
train_size = int(len(df) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

#Train the Linear Regression Model
model = LinearRegression()
model.fit(X_train, y_train)

#Predict on the Test Set and Evaluate Model Performance
y_pred = model.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("Test RMSE:", rmse)
```

```python
#Predict the Next 30 Days Using the Linear Trend from the Model
last_index = X[-1, 0]  # Last index in the original dataset
future_indices = np.array([last_index + i + 1 for i in range(30)]).reshape(-1, 1)
future_predictions = model.predict(future_indices)
```

```python
#Visualize the Actual vs. Predicted for Test Data and Future Prediction

# Plot the Test Data predictions
plt.figure(figsize=(12, 6))
plt.plot(df['date'].iloc[train_size:], y_test, label='Actual Close Prices', color='blue')
plt.plot(df['date'].iloc[train_size:], y_pred, label='Predicted Close Prices (Test)', color='orange')
plt.title('Actual vs. Predicted Close Prices (Test Data)')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()

```

```python

# Ensure the 'date' column is in datetime format
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# Plot the Test Data predictions 
plt.figure(figsize=(12, 6))
plt.plot(df['date'].iloc[train_size:], y_test, label='Actual Close Prices', color='blue')
plt.plot(df['date'].iloc[train_size:], y_pred, label='Predicted Close Prices (Test)', color='orange')

# Set the title and labels
plt.title('Actual vs. Predicted Close Prices (Test Data)')
plt.xlabel('Date')
plt.ylabel('Price')

# Rotate and format x-axis dates for readability
plt.xticks(rotation=45)  # Rotate x-axis labels by 45 degrees
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))  # Format as 'Year-Month'

# set major ticks at monthly intervals if your dates are dense
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())

# Add the legend and show the plot
plt.legend()
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()

```

```python
df['date'] = pd.to_datetime(df['date'])

# Select the last 450 days of data(to make a more focused plot)
last_450_days = df[-450:]  # Last 450 days of actual data

# Extend the date range for the next 30 days from the last date in the dataset
last_date = df['date'].iloc[-1]  # Now this will be a datetime object
future_dates = pd.date_range(last_date + timedelta(days=1), periods=30)

# Plot the Last 450 Days and the Next 30 Days Predictions
plt.figure(figsize=(12, 6))
plt.plot(last_450_days['date'], last_450_days['close'], label="Last 450 Days Actual Price", color="blue")
plt.plot(future_dates, future_predictions, label="Next 30 Days Prediction", color="orange")
plt.title("Bitcoin Price: Last 450 Days and Next 30 Days Prediction")
plt.xlabel("Date")
plt.ylabel("Bitcoin Price")
plt.legend()
plt.show()
```
