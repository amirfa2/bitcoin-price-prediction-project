```python
import requests
import pandas as pd
# Set up the API endpoint and parameters
url = "https://min-api.cryptocompare.com/data/v2/histoday"
params = {
    'fsym': 'BTC',  # From symbol (BTC)
    'tsym': 'USD',  # To symbol (USD)
    'limit': 2000,  # Get data for 2000 days
    'aggregate': 1,  # Daily data
}
# Send the request
response = requests.get(url, params=params)
data = response.json()
# Extract the relevant data
df_2 = pd.DataFrame(data['Data']['Data'])
# Convert the timestamp to a readable date format
df_2['time'] = pd.to_datetime(df_2['time'], unit='s')
df_2.head()
```

```python
# putting df in a shape that we need
df_2.rename(columns={'time': 'date'}, inplace=True)
```

```python
df_2.rename(columns={'volumefrom': 'volume'}, inplace=True)
df_2.rename(columns={'volumeto': 'total_value'}, inplace=True)
df_1.rename(columns={'Volume USD': 'total_value'}, inplace=True)
df_1.rename(columns={'Volume BTC': 'volume'}, inplace=True)
```

```python
# converting to datetime format
df_1['date'] = pd.to_datetime(df_1['date'])
df_2['date'] = pd.to_datetime(df_2['date'])
```

```python
# setting the date as an index
df_1.set_index('date', inplace=True)
df_2.set_index('date', inplace=True)
```

```python
# concatinating 2 df into one
df = pd.concat([df_1, df_2])
df.head()
```

```python
# sorting dt by date
df.sort_index(inplace=True)
```

```python
df.head()
```

```python
# create a date range covering all days from the earliest to the latest date
date_range = pd.date_range(start=df.index.min(), end=df.index.max(), freq='D')
date_range
```

```python
# reindex the df to ensure there is one row for every day in the range
df = df[~df.index.duplicated(keep='first')]
df = df.reindex(date_range)
```

```python
# dropping columns that we won't use
df.drop(columns=["conversionType", "conversionSymbol"], inplace=True)
```

```python
# dropping columns that we won't use
df.drop(columns=["unix", "symbol"], inplace=True)
```

```python
# setting index back
df = df.reset_index()
df
```

```python
# rename the date
df.rename(columns={'index': 'date'}, inplace=True)
```

```python
# Calculate Weighted Price
df['weighted_price'] = df['total_value'] / df['volume']
```

```python

pre_df = df[['date', 'weighted_price']]
pre_df.set_index('date', inplace=True)
pre_df
```

```python
weekly_df = pre_df.resample('W').mean()
```

```python
plt.figure(figsize=(12, 6))  # Set the figure size
plt.plot(weekly_df, color='green', lw=2)  # Plot the weekly data
plt.title("Weekly Bitcoin Price", fontsize=17)
plt.xlabel('Time in Weeks', fontsize=17)
plt.ylabel('$(Dollar)', fontsize=17)
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.grid(True)  # Add grid for easier reading
plt.show()
```
